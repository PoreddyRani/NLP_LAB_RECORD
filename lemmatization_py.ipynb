```python
import spacy

# Download the model first if you haven't:
# python -m spacy download en_core_web_sm

# Load English model
nlp = spacy.load("en_core_web_sm")

text = "The striped bats are hanging on their feet and ate better than before."

doc = nlp(text)

print("Token : Lemma")
for token in doc:
    print(f"{token.text} : {token.lemma_}")
```
