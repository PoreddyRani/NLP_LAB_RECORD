```python
import re
def normalize_text(text):
  text=text.lower()
  text=re.sub(r'[^\w\s]','',text)
  text=re.sub(r'\s+','',text).strip()
  return text
raw="hello, world !! this is python."
print(normalize_text(raw))
```
