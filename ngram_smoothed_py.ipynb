```python
from collections import Counter
text = "I love NLP and I love Python".split()
bigrams = [(text[i], text[i+1]) for i in range(len(text)-1)]
V = len(set(text))                       # Vocabulary size
count_uni = Counter(text)
count_bi = Counter(bigrams)
def smooth_prob(w1, w2): return (count_bi[(w1, w2)] + 1) / (count_uni[w1] + V)
print("Smoothed P(love|I) =", smooth_prob("I", "love"))
```
