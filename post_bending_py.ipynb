```python
import spacy

# Load the small English model
nlp = spacy.load("en_core_web_sm")

# Input sentence
text = "The quick brown fox jumps over the lazy dog."

# Process the text
doc = nlp(text)

# Print word and its POS tag
for token in doc:
    print(token.text, "→", token.pos_)
```
