```python
import spacy

# Load the English NLP model
nlp = spacy.load("en_core_web_sm")

# Sample text
text = "The cats are running swiftly."

# Process the text
doc = nlp(text)

# Extract and display morphological features
for token in doc:
    print(f"Word: {token.text}")
    print(f"  Lemma: {token.lemma_}")
    print(f"  POS: {token.pos_}")
    print(f"  Morph: {token.morph}")
    print("-" * 40)
```
