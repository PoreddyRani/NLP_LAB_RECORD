```python
from nltk.tag import hmm

train = [[('The','DT'),('dog','NN'),('barks','VBZ')],
         [('A','DT'),('cat','NN'),('meows','VBZ')]]

tagger = hmm.HiddenMarkovModelTrainer().train_supervised(train)
print(tagger.tag(['A','dog','meows']))
```
