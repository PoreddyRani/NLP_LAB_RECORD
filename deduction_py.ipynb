```python
premises = ["rain -> wet", "rain"]
goal = "wet"

if "rain -> wet" in premises and "rain" in premises:
    result = True
else:
    result = False

print("Premises:", premises)
print("Hypothesis:", goal)
print("Can we deduce it?", result)
```
